```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

forestdataC = '/content/drive/MyDrive/content/forestdata.csv'
forestdataC = pd.read_csv(forestdataC)
forestdataC = forestdataC.iloc[:, :-3]
forestdataC = forestdataC.iloc[:-1, :]
forestdataC.replace("...", pd.NA, inplace=True)
forestdataC = forestdataC.dropna(axis=0, how='any')


miningvalueC = '/content/drive/MyDrive/content/miningvalueT.csv'
miningvalueC = pd.read_csv(miningvalueC)
miningvalueC.replace("...", pd.NA, inplace=True)
miningvalueC = miningvalueC.dropna(axis=0, how='any')

# long forma
forestdata_com = forestdataC.melt(id_vars='Country and Area', var_name='year', value_name='deforestation')
forestdata_com.columns = ['country', 'year', 'deforestation']
miningdata_com = miningvalueC.melt(id_vars='Country and Area', var_name='year', value_name='mining')
miningdata_com.columns = ['country', 'year', 'mining']



forestdata_com['country'] = forestdata_com['country'].str.strip()
miningdata_com['country'] = miningdata_com['country'].str.strip()
forestdata_com['year'] = forestdata_com['year'].astype(str).str.strip()
miningdata_com['year'] = miningdata_com['year'].astype(str).str.strip()



merged_long_df = pd.merge(forestdata_com, miningdata_com, on=['country', 'year'], how='outer')
print("merged_long_df:")
print(merged_long_df.head())





G = nx.Graph()

for index, row in merged_long_df.iterrows():
    if pd.notnull(row['deforestation']):
        G.add_edge(row['country'], f"Deforestation_{row['year']}", type='deforestation')
    if pd.notnull(row['mining']):
        G.add_edge(row['country'], f"Mining_{row['year']}", type='mining')


plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_size=500, node_color='lightblue', font_size=10)
plt.title("Country–Year Deforestation & Mining")
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# mkaeing valsnumeric
merged_long_df['deforestation'] = pd.to_numeric(merged_long_df['deforestation'], errors='coerce')
merged_long_df_clean = merged_long_df.dropna(subset=['deforestation'])

# pivot countries rows, years columns
pivot_df = merged_long_df_clean.pivot_table(index='country', columns='year', values='deforestation')
pivot_df = pivot_df.dropna()





if pivot_df.empty: #kept being empty
    print("empty.")
else:
    # KMeans
    kmeans = KMeans(n_clusters=4, random_state=42)
    pivot_df['Cluster'] = kmeans.fit_predict(pivot_df)

    # PCA heard of this before and googled it, looks interesting
    pca = PCA(n_components=2)
    components = pca.fit_transform(pivot_df.drop(columns='Cluster'))

    pca_df = pd.DataFrame(components, columns=['PC1', 'PC2'])
    pca_df['Cluster'] = pivot_df['Cluster'].values
    pca_df['Country'] = pivot_df.index


    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='Set2', s=100) #didn't know this was the format
    plt.title('KMeans Clustering by Deforestation')
    plt.xlabel('Part 1')
    plt.ylabel('Part 2')
    plt.grid(True)
    plt.legend(title='Cluster')
    plt.show()

    # shows the countries in each cluster
    for i in range(4):
        print(f"\n Cluster {i}:")
        print(", ".join(pivot_df[pivot_df['Cluster'] == i].index)) #this was completely wrign earlier
```

```python
import pandas as pd
import matplotlib.pyplot as plt


valid_years = ['1990', '1995', '2000', '2005', '2010', '2015', '2020'] #year was still messy had to call specfic columns
merged_long_df = merged_long_df[merged_long_df['year'].isin(valid_years)]


merged_long_df['deforestation'] = pd.to_numeric(merged_long_df['deforestation'], errors='coerce')
merged_long_df['mining'] = pd.to_numeric(merged_long_df['mining'], errors='coerce')

#tyring to plot again
def plot_trend(country):
    df = merged_long_df[merged_long_df['country'] == country].copy()

    df['year'] = df['year'].astype(int)
    df.sort_values('year', inplace=True)


    df.dropna(subset=['deforestation', 'mining'], how='all', inplace=True) # dropping again sicne errors came up





    if df['deforestation'].notna().sum() > 0:
        plt.plot(df['year'], df['deforestation'], label='Deforestation')
    if df['mining'].notna().sum() > 0:
        plt.plot(df['year'], df['mining'], label='Mining')

    plt.title(f"{country} – Deforestation vs Mining")
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Step 4: Try it out!
plot_trend('Brazil')
```

```python
def plot_trend_dual_axis(country):
    df = merged_long_df[merged_long_df['country'] == country].copy()
    df['year'] = df['year'].astype(int)
    df.sort_values('year', inplace=True)
    df.dropna(subset=['deforestation', 'mining'], how='all', inplace=True)

    fig, ax1 = plt.subplots()

    #  deforestation
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Deforestation (sq km)', color='tab:blue')
    ax1.plot(df['year'], df['deforestation'], label='Deforestation', color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    #  mining
    ax2 = ax1.twinx()
    ax2.set_ylabel('Mining Value', color='tab:orange')
    ax2.plot(df['year'], df['mining'], label='Mining', color='tab:orange')
    ax2.tick_params(axis='y', labelcolor='tab:orange')

    plt.title(f"{country} – Deforestation vs Mining (Dual Axis)")
    fig.tight_layout()
    plt.show()


plot_trend_dual_axis('Brazil')
```
